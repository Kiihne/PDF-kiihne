```python
#comparing transiton points for spline and piecewise column densities
#slopes plot extracted from txt files
#slopes plot for 3D plots: extract 3D volume slopes from npz, convert 2D column slopes to 3D
import numpy as np
import yt
import time
from yt.funcs import mylog
import matplotlib.pyplot as plt
mylog.setLevel(40) #prevents warning
plt.style.use('default')
import scipy
from scipy import optimize
import pandas as pd
plt.rcParams["font.family"] = "serif"
```

```python
#main plotting cell
line = ('solid', 'dashed', 'dotted')
#inputfile = ('txt_file_storage/slope_0%_lim,5.txt','txt_file_storage/slope_1%_lim,5.txt','txt_file_storage/slope_2%_lim,5.txt','txt_file_storage/slope_3%_lim,5.txt','txt_file_storage/slope_4%_lim,5.txt','txt_file_storage/slope_5%_lim,5.txt')#,'slope_10%_and_6%.txt',)
inputfile = ('txt_file_storage/trans_density0%.txt','txt_file_storage/trans_density1%.txt','txt_file_storage/trans_density2%.txt','txt_file_storage/trans_density3%.txt','txt_file_storage/trans_density4%.txt','txt_file_storage/trans_density5%.txt')

plt.figure(figsize = (4.5,4.5))
trans_g = []
trans_gt = []
trans_gtm = []
trans_gtmp = []
for a in range(3):

    sfe = range(6)
    for ii in range(6):
        #slope cutting. data from txt file chopped into usable form.
        f =open(inputfile[ii])
        newlist = [i for i in f]
        string = newlist[0]
        string1 = string.split("[")[1]
        string2 = string1.split("]")[0]
        string3 = string2.split(",")
        string3 = np.array(string3)
        slope_list = string3.astype('float')
        trans_g.append(slope_list[0+a])
        trans_gt.append(slope_list[3+a])
        trans_gtm.append(slope_list[6+a])
        trans_gtmp.append(slope_list[9+a])
     
    
inputfile = ('txt_file_storage/trans_point0%.txt','txt_file_storage/trans_point1%.txt','txt_file_storage/trans_point2%.txt','txt_file_storage/trans_point3%.txt','txt_file_storage/trans_point4%.txt','txt_file_storage/trans_point5%.txt')
trans_spline_g = []
trans_spline_gt = []
trans_spline_gtm = []
trans_spline_gtmp = []
for a in range(3):

    sfe = range(6)
    for ii in range(6):
        #slope cutting. data from txt file chopped into usable form.
        f =open(inputfile[ii])
        newlist = [i for i in f]
        string = newlist[0]
        string1 = string.split("[")[1]
        string2 = string1.split("]")[0]
        string3 = string2.split(",")
        string3 = np.array(string3)
        slope_list = string3.astype('float')
        trans_spline_g.append(slope_list[0+a])
        trans_spline_gt.append(slope_list[3+a])
        trans_spline_gtm.append(slope_list[6+a])
        trans_spline_gtmp.append(slope_list[9+a])
    x=.1*np.linspace(-10,35)
    y=x
plt.plot(x,y, color = 'black')
plt.scatter(trans_spline_gt,trans_gt, marker = "+", color = 'C2', label = 'Turbulence')#, alpha =alpha[a])
plt.scatter(trans_spline_gtm, trans_gtm, marker = "s", color = 'C1', label = 'B-Fields')#, alpha =alpha[a])
plt.scatter(trans_spline_gtmp, trans_gtmp, marker = "o", color = 'C0', label = 'All + Outflows')#, alpha =alpha[a])
plt.title('spline', size = 15)
plt.xlabel(r'spline transition density', size=11)
plt.xlim(-.2,3.25)
plt.ylim(-.2,3.25)
plt.ylabel(r'piecewise transition density', size=11)
plt.legend()

plt.tight_layout()

plt.savefig('trans_density_compare')
plt.show
```
